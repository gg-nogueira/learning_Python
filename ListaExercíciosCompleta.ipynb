```python
import numpy as np
import random
```

```python
#Exercício 1

agenda={}

while True:
  x=int(input("1-Mostrar contato\n2- Adicionar Contato\n3- Deletar contato\n4- Sair\nDigite sua opção:\n"))
  if x==1:
    display(agenda)
    print()
  elif x==2:
    nome=input("Qual o nome do contato? ")
    num=input("Qual o número do contato? ")
    agenda[nome]=num
  elif x == 3:
    nome_apaga = input("Contato de quem você deseja apagar? ")
    if nome_apaga in agenda:
        nome_apagado = agenda.pop(nome_apaga)
        print(f"O contato {nome_apagado} foi apagado")
    else:
      print("O nome inserido não existe na agenda. No menu, pressione 1 para consultá-la")
  elif x==4:
    print("Obrigado por acessar!")
    break
  else:
    print("Número digitado inválido")
    

```

```python
#Exercício 4

x=int(input(f"Qual o tamanho "))
vet = np.array([random.randint(0,50) for i in range(x)])
maior = vet[0]

for i in range(0, x):
    if vet[i] > maior:
        maior = vet[i]

print(f"Vetor original: {vet}")
print(f"Maior valor: {maior}")
```

```python
#Exercício 5

x=int(input(f"Qual o tamanho "))
vet = np.array([random.randint(0,50) for i in range(x)])

soma=sum(vet)

print(f"Vetor original: {vet}")
print(f"Média: {soma/len(vet)}")


```

```python
#Exercício 6

x=int(input(f"Quantos cadastros serão feitos? "))
cadastros = np.empty(x, dtype=object)

for i in range (x):
  nome=input("Qual o nome? ")
  idade=int(input("Qual idade? "))
  lista = {"nome": nome, "idade": idade}
  cadastros[i]=lista

mais_velha = cadastros[0]

for pessoa in cadastros:
    if pessoa["idade"] > mais_velha["idade"]:
        mais_velha = pessoa


print(f"Pessoa mais velha: {mais_velha['nome']}, {mais_velha['idade']} anos")
```
