# Sobre tuplas e listas

```python
# A função nativa "sorted" transmorma o item em lista e ordena os valores
# Pode utilizar também para ordem decrescente.
print('Declarando tupla e lista e imprimindo em ordem os valores.')
a = (2, 1, 3)
b = [2, 4, 3]

print(sorted(a)) # Imprime como lista
print(sorted(b, reverse=True))

# Veja que a função "sorted" não altera as variáveis.
print('\nOs valores originais não são alterados.')
print(a)
print(b)

# Ela retorna uma nova lista e pode ser armazenada na mesma variável ou em outra
print('\nAtribuindo "sorted" de "a" nele mesmo e exibindo.')
a = sorted(a)
print(a)
```

```python
# Uma tupla necessariamente deve ser declarada com parênteses e deve ter
# ao menos 1 vírgula (,)
print('Declarando variáveis e exibindo o tipo de dado armazenado.')
c = (10)
d = [20]

print(type(c), type(d))

# Criando uma tupla em "c"
print('\nAjustando a declaração de "c".')
c = (10,) # ou tuple([10]) - converte lista em tupla
print(c, d)
print(type(c), type(d))
```

```python
# Uma lista pode ser manipulada, diferente de tupla
print('Declarando lista "e".')
e = [i*2+i-1 for i in range(10)]
print(e)

print('\nExibindo insert.')
# Insere no índice 1 o valor 0
e.insert(1, 0)
print(e)

print('\nExibindo pop.')
# Remove de "e" o valor do índice 1 e armazena o valor em "f"
f = e.pop(1) # Se não passar o índice, remove o último valor
print(e, 'valor removido:', f)


print('\nExibindo append.')
# Semelhante ao "insert", porém sempre adciona o valor no fim da lista
e.append(29)
print(e)
```
