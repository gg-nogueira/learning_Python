# Atividade 1

```python
while True:
  try:
    n1 = int(input('Digite um número: '))
    n2 = int(input('Digite outro número: '))

    # Menu
    print('1 - Somar\n2 - Subtrair\n3 - Dividir\n4 - Multiplicar\n0 - Sair\n')
    aux = int(input('O que deseja fazer? '))
    
    if aux == 0:
      break
    elif aux == 1:
      res = n1 + n2
    elif aux == 2:
      res = n1 - n2
    elif aux == 3:
      res = n1 / n2
    elif aux == 4:
      res = n1 * n2
    else:
      print('Valor errado, moscão')
      pass
    
    print(res)

  except ValueError:
    print('Valor inválido!')
  except ZeroDivisionError:
    print('Não pode dividir por zero moscão!')

n=3
print('Saindo.... '+('\n'*n)+'Falou moscão!')
```

```python
def verify(n=[]):
  for i in n:
    if i > 10 and i < 0:
      return True

while True:
  try:
    n1 = float(input('Digite nota 1: '))
    n2 = float(input('Digite nota 2: '))
    n3 = float(input('Digite nota 3: '))

    notas = [n1, n2, n3]
    if verify(notas):
      print('Nota inválida!')
    else:
      media = (n1+n2+n3)/3.0

      pres = float(input('Das 20 aulas, quantas aulas você veio? '))
      presp = (pres/20)*100
      print(presp, media)
      if (media >= 6) and (presp >=75):
        print('passou uau')
      else:
        print('putz pegou dp burrao')
      break
  except ValueError:
    print('Valor inválido!')
```
