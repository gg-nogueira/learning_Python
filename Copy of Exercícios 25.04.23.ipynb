1. Crie uma agenda de contatos contendo nome e celular de n pessoas, armazenados em um dicionário. Em seguida crie métodos para consultar, inserir e deletar itens neste dicionário. Os dados devem ser lidos do teclado.

```python
# Iniciando o dicionário "agenda" com valores de chaves em listas vazias.
agenda = {
    'nome': [],
    'celular': []
}

# Função para mostrar a lista de contatos.
def ver_agenda():

  # Se houver um contato salvo exibe a lista.
  if len(agenda['nome']) > 0:
    print() # Para pular uma linha.

    # Usamos o "zip" para obter os valores de duas listas ao mesmo tempo
    for i, j in zip(agenda['nome'], agenda['celular']):
      print(f'{i} - {j}')
      print()
  
  # Se não houver contatos, exibe a mensagem abaixo.
  else:
    print('\nNão há nenhum contato salvo aqui!')

# Função para adicionar um contato.
def adicionar():
  print('\nAdicionar Contato\n')

  # Recebe nome e número do teclado.
  nome = input('Nome: ')
  numero = input('Número: ')

  # Verifica se os dados inseridos são válidos.
  # Se sim, adiciona nas listas do dicionário.
  if nome != '' and numero != '':
    agenda['nome'].append(nome)
    agenda['celular'].append(numero)
  else:
    print('\nNome ou celular inválidos.')

# Função bônus.
# Modifica um contato.
def modificar():

  # Inicia com tratativa de erro.
  try:

    # Exibe um menu, para saber o que vamos modificar.
    print('\nModificar Contato\n')
    print('1 - Modificar nome\n2 - Modificar número')

    # Recebe a opção desejada via teclado.
    opcao = int(input('\nEscolha uma das opções acima: '))
    print() # Apenas para pular linha.
    
    # Modificando o nome.
    if opcao == 1:

      # Exibe a lista de contatos com os nomes e índice para escolha.
      for i, j in enumerate(agenda['nome']):
        print(f'{i} - {j}')
      
      # Sobrepõe o valor de "opcao" com o valor do menu exibido acima.
      opcao = int(input('\nEscolha um nome acima: '))

      # Exibe o nome a ser alterado e solicita o novo valor.
      print(f'\nAlterando {agenda["nome"][opcao]}.')
      agenda['nome'][opcao] = input('Insira o novo nome: ')

    # Modificando o número.
    elif opcao == 2:

      # Exibe a lista de contatos com os números e índice para escolha.
      for i, j in enumerate(agenda['celular']):
        print(f'{i} - {j}')
      
      # Sobrepõe o valor de "opcao" com o valor do menu exibido acima.
      opcao = int(input('Escolha um celular acima: '))

      # Exibe o número a ser alterado e solicita o novo valor.
      print(f'\nAlterando {agenda["celular"][opcao]}.')
      agenda['celular'][opcao] = input('Insira o novo número: ')
    
    # Caso a opção selecionada não esteja no menu.
    else:
      print('\nOpção inválida.')
  
  # Tratativa de erro para valor inválido (Caso o número de celular fosse do tipo numérico).
  except ValueError:
    print('\nInformação inválida.')

  # Tratativa de erro para opção de menu inválida.
  except IndexError:
    print('\nOpção de nome ou celular inválida.')

# Função para deletar contato.
def deletar():
  print('\nDeletar Contato\n')

  # Exibe um índice, nome e número, nesta ordem, em tabela.
  for i, j in enumerate(zip(agenda['nome'], agenda['celular'])):
    print(f'{i} - {j[0]} ({j[1]})')

  # Recebe o valor excolhido do menu acima.
  opcao = int(input('\nEscolha um contato acima: '))

  # Pede confirmação para excluir o contato.
  print(f'\nDeseja mesmo excluir {agenda["nome"][opcao]} ({agenda["celular"][opcao]}) ?')
  print('1 - Sim\n[n] - Digite qualquer outro valor para cancelar.')
  confirma = int(input('\nEscolha uma opção acima: '))

  # Se confirmado exclui o contato
  if confirma == 1:
    agenda['nome'].pop(opcao)
    agenda['celular'].pop(opcao)
    print('\nContato excluido!')

  # Se não, cancela toda a operação e retorna ao menu principal.
  else:
    print('\nCancelado!')

# Função do menu principal.
def menu():

  # Inicia com tratativa de erro
  try:

    # Inserindo as funções em uma lista.
    funcoes = [ver_agenda, adicionar, modificar, deletar]

    # Exibe o menu.
    print('\n1 - Ver agenda')
    print('2 - Adicionar contato')
    print('3 - Modificar contato') # Bônus
    print('4 - Deletar contato')
    print('5 - Sair')

    # Recebe o valor do menu e subtrai 1.
    opcao = int(input('\nEscolha uma opção do menu: ')) - 1
    
    # Opção para sair do programa.
    if opcao == 4:
      print('\nObrigado por usar nosso app!')

      # Retorna False para parar o laço de repetição "while" no bloco de código abaixo.
      return False
    else:
      # Caso escolha uma opção do menu, ele recebe a função no índice da opção
      # e executa a função.
      executa = funcoes[opcao]
      executa()
    
    # Retorna sempre True caso o programa não seja encerrado.
    return True
  
  # Tratativa de erro caso a opção escolhida não esteja dento da lista de funções.
  except IndexError:
    print('\nOpção inválida.')
    return True
  
  # Tratativa de erro caso o valor seja não númerico.
  except ValueError:
    print('\nValor inválido. Insira apenas números.')
```

```python
# Executa o prgrama do bloco acima
# E só encerra caso seja selecionado a opção "Sair" do menu principal.
run = True
while run:
  run = menu()
```

2. Faça um programa que carregue dados em um dicionário, onde a chave deverá ser um número aleatório entre 0 e 100 e o valor deverá ser uma palavra pré-cadastrada em uma tupla.

```python
import random

# Declarando uma tupla com nomes e um dicionário vazio.
pessoas = ('João', 'Maria', 'José', 'Pedro', 'Ana')
dicio = {}

# Adicionando chaves aleatórias ao dicionário com os valores da tupla.
for i in pessoas:

  # Bloco para adiconar chaves aleatórias que não se repetem.
  while True:

    # Valor aleatório entre 0 e 100.
    a = random.randint(0, 100)

    # Verifica se o valor aleatório não está no dicionário.
    # Se sim, ele adiciona o nome no dicionário e para o laço "while".
    # Caso já esteja, não faz nada e repete "while".
    if a not in dicio.keys():
      dicio[a] = i
      break

# Para confirmar que não se repete nenhum valor aleatório,
# Teste modificar os valores aleatórios entre 0 e 4.
print(dicio)
```

3. Cadastre 5 funcionários em um dicionário, pelo teclado, com as seguintes chaves: nome, idade, cargo, salário. No final, mostre o conteúdo do dicionário e a média dos salários.

```python
# Iniciando com a tratativa de erro.
try:
  
  # Gerando um dicionário com listas vazias como valor de cada chave.
  fun = {
      'Nome': [],
      'Idade': [],
      'Cargo': [],
      'Salario': []
  }
  
  # Variável delimitar a quantidade de funcionarios a serem cadastrados.
  n_fun = 2

  # Acumulador para os salários dos funcionários.
  total_salario = 0

  # Obtendo os dados via teclado.
  for i in range(n_fun):
    print(f'\nFuncionário {i+1}')
    fun['Nome'].append(input('Nome: '))
    fun['Idade'].append(int(input('Idade: ')))
    fun['Cargo'].append(input('Cargo: '))
    fun['Salario'].append(float(input('Salário: ')))

    # Acumulando os salários.
    total_salario += fun['Salario'][i]

  # Opção de exibição 1 (Simples).
  print('\nExibição 1')
  display(fun)
  
  # Opção de exibição 2 (Mais bonita - Em tabela).
  print(f'\n{"Nome":{" "}<{15}} {"Idade":{" "}<{8}} {"Cargo":{" "}<{15}} {"Salário":{" "}<{10}}')
  for nome, idade, cargo, salario in zip(fun['Nome'], fun['Idade'], fun['Cargo'], fun['Salario']):
    print(f'{nome:{" "}<{15}} {str(idade):{" "}<{8}} {cargo:{" "}<{15}} R${str(salario):{" "}<{10}}')
  
  # Dividindo o somatório de salários pela quantidade de funcionários. (Média).
  total_salario /= n_fun
  print(f'\nA média dos salários é R${total_salario:.2f}')

# Tratativa de erro caso valor de idade e salário não sejam numéricos.
except ValueError:
  print('\nDado inserido é inválido.')
```

4. Carregue um vetor com 5 números inteiros e mostre o maior deles, sem usar a função max().

```python
import numpy as np
import random

# Gerando uma lista com valores aleatórios entre 1 e 50 e passando para vetor.
v = np.array([random.randint(1, 50) for i in range(5)])

# Opção 1 - Função que retorna o maior valor de uma lista ou vetor.
def maior(ls):
  maior = ls[0]
  for i in ls:
    if i > maior:
      maior = i
  return maior

# Opção 2 - Ordenar os valores em ordem crescente e obter apenas o último.
maior_valor = np.sort(v)[-1]

# Exibindo os valores.
print('Vetor: ', v)
print('Opção 1: ', maior(v))
print('Opção 2:', maior_valor)
```

5. Carregue um vetor com 3 notas e mostre a média das notas. (Sem usar a função mean())

```python
import numpy as np
import random

# Gerando uma lista com valores aleatórios entre 0 e 10 e passando para vetor.
notas = np.array([random.randint(0, 10) for i in range(3)])

# Função que retorna a média de uma lista ou vetor.
def media(ls):
  return float(sum(ls) / len(ls))

# Atribuindo o valor retornado da função à "minha_media".
minha_media = media(notas)

# Exibindo as notas e a média.
print(f'Notas: {notas}')
print(f'Média: {minha_media:.2f}')
```
